```python
import numpy as np
import tensorflow as tf

from tensorflow import keras
```

```python
boston_housing = keras.datasets.boston_housing
```

```python
(train_data,train_output),(test_data, test_output) = boston_housing.load_data()
```

```python
x_values = train_data[:,[0]]
```

```python
# x values from the training dataset
x = tf.placeholder(tf.float32,[None,1])
#actual prediction values of influencial parameter
y1 = tf.placeholder(tf.float32,[None,1])
```

```python
# w as a dependent variable depending on x as well as y
w = tf.Variable(tf.zeros([1,1]))
# b only depends upon x
b = tf.Variable(tf.zeros([1]))
```

```python
y = tf.add(tf.matmul(x,w),b)
```

```python
cost_func = tf.reduce_sum(tf.pow((y-y1),2))
```

```python
train_step = tf.train.GradientDescentOptimizer(0.00000001).minimize(cost_func)
```

```python
init = tf.initialize_all_variables()
```

```python
sess = tf.Session()
sess.run(init)
```

```python
feed = {x: x_values, y1: train_output}
```

```python
sess.run(train_step,feed_dict = feed)
```
